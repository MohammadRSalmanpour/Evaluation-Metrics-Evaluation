# Statistical Test

```python
data1 = x
data2 = y
result = {}
```

# **Scipy Library:**

```python
# Independent t-test
itt_statistic, itt_p_value = stats.ttest_ind(data1, data2)

# Paired t-test
ptt_statistic, ptt_p_value = stats.ttest_rel(data1, data2)

# Kolmogorov-Smirnov Test
kst_statistic, kst_p_value = stats.ks_2samp(data1, data2)

# ANOVA
ANOVA_f_statistic, ANOVA_p_value = stats.f_oneway(data1, data2)

# Kruskal-Wallis Test
kwt_statistic, kwt_p_value = stats.kruskal(data1, data2)

# Mann-Whitney U Test
mwut_statistic, mwut_p_value = stats.mannwhitneyu(data1, data2)

# Shapiro-Wilk Test
swt_statistic, swt_p_value = stats.shapiro(data1)

# F-test
F_statistic, F_p_value = stats.f_oneway(data1, data2)

# Permutation test
perm_result = stats.permutation_test((data1, data2), lambda x, y: np.mean(x) - np.mean(y))

# Welch's t-test
wet_statistic, wet_p_value = stats.ttest_ind(data1, data2, equal_var=False)

# Bartlett's test
bt_statistic, bt_p_value = stats.bartlett(data1, data2)

# Levene's test
lt_statistic, lt_p_value = stats.levene(data1, data2)

# Wilcoxon signed-rank test
wil_statistic, wil_p_value = stats.wilcoxon(data1, data2)

# Chi-Square Test
observed = np.array([30, 10, 40, 20])
expected = np.array([25, 25, 25, 25])
chi2_statistic, chi2_p_value = stats.chisquare(observed, f_exp=expected)

result['Scipy'] = {
    'itt_statistic': itt_statistic,
    'itt_p_value': itt_p_value,
    'ptt_statistic': ptt_statistic,
    'ptt_p_value': ptt_p_value,
    'kst_statistic': kst_statistic,
    'kst_p_value': kst_p_value,
    'chi2_statistic': chi2_statistic,
    'chi2_p_value': chi2_p_value,
    'ANOVA_F_statistic': ANOVA_f_statistic,
    'ANOVA_p_value': ANOVA_p_value,
    'kwt_statistic': kwt_statistic,
    'kwt_p_value': kwt_p_value,
    'mwut_statistic': mwut_statistic,
    'mwut_p_value': mwut_p_value,
    'swt_statistic': swt_statistic,
    'swt_p_value': swt_p_value,
    'F_statistic': F_statistic,
    'F_p_value': F_p_value,
    'pt_statistic': perm_result.statistic,
    'pt_p_value': perm_result.pvalue,
    'wet_statistic': wet_statistic,
    'wet_p_value': wet_p_value,
    'bt_statistic': bt_statistic,
    'bt_p_value': bt_p_value,
    'lt_statistic': lt_statistic,
    'lt_p_value': lt_p_value,
    'wsrt_statistic': wil_statistic,
    'wsrt_p_value': wil_p_value
}
```

# **Lifelines Library:**

```python
# Log-Rank Test
durations = np.random.randint(1, 100, size=100)
event_observed = np.random.randint(0, 2, size=100)
results_logrank = logrank_test(durations, durations, event_observed_A=event_observed, event_observed_B=event_observed)

result['Lifelines'] = {
    'lrt_statistic': results_logrank.test_statistic,
    'lrt_p_value': results_logrank.p_value
}
```

# **Statsmodels Library:**

```python
# Z-test
z_statistic, z_p_value = ztest(data1, data2)

# Independent t-test
cm = CompareMeans(DescrStatsW(data1), DescrStatsW(data2))
itt_statistic_sm, itt_p_value_sm, _ = cm.ttest_ind()

# Paired t-test (Scipy)
d = data1 - data2
cm = sm.stats.DescrStatsW(d)
ptt_statistic_sm, ptt_p_value_sm, _ = cm.ttest_mean(0)
ecdf1 = sm.distributions.ECDF(data1)
ecdf2 = sm.distributions.ECDF(data2)
x = np.sort(np.concatenate([data1, data2]))
cdf1 = ecdf1(x)
cdf2 = ecdf2(x)

# Calculate the KS statistic
kst_statistic_sm2 = np.max(np.abs(cdf1 - cdf2))

# Calculate the p-value using SciPy
kst_statistic_sm2, kst_p_value_sm2 = stats.ks_2samp(data1, data2)


data = pd.DataFrame({
    'value': np.concatenate([data1, data2]),
    'group': ['data1'] * len(data1) + ['data2'] * len(data2)
})

#ANOVA
model = smf.ols('value ~ group', data=data).fit()
anova_table_sm = sm.stats.anova_lm(model, typ=2)
anova_f_statistic_sm = anova_table_sm['F'][0]
anova_p_value_sm = anova_table_sm['PR(>F)'][0]

# Welch's t-test
wet_statistic_sm, wet_p_value_sm, df = ttest_ind(data1, data2, usevar='unequal')

result['Statsmodels'] = {
    'z_statistic': z_statistic,
    'z_p_value': z_p_value,
    'itt_statistic': itt_statistic_sm,
    'itt_p_value': itt_p_value_sm,
    'ptt_statistic': ptt_statistic_sm,
    'ptt_p_value': ptt_p_value_sm,
    'kst_statistic': kst_statistic_sm2,
    'kst_p_value': kst_p_value_sm2,
    'ANOVA_F_statistic': anova_f_statistic_sm,
    'ANOVA_p_value': anova_p_value_sm,
    'wet_statistic': wet_statistic_sm,
    'wet_p_value': wet_p_value_sm,
}
```

# **Pingouin Library:**

```python
itt_pg = pg.ttest(data1, data2, paired=False)
ptt_pg = pg.ttest(data1, data2, paired=True)
data_pg = pd.DataFrame({'data': np.concatenate([data1, data2]), 'group': ['data1'] * len(data1) + ['data2'] * len(data2)})
ANOVA_pg = pg.anova(dv='data', between='group', data=data_pg)
kwt_pg = pg.kruskal(data_pg, dv='data', between='group')
mwut_pg = pg.mwu(data1, data2)
swt_pg = pg.normality(data1)
wet_pg = pg.ttest(data1, data2, paired=False, correction=True)
zt_pg = pg.ttest(data1, data2, paired=False, alternative='two-sided')

result['Pingouin'] = {
    'itt_statistic': itt_pg['T'][0],
    'itt_p_value': itt_pg['p-val'][0],
    'ptt_statistic': ptt_pg['T'][0],
    'ptt_p_value': ptt_pg['p-val'][0],
    'ANOVA_F_statistic': ANOVA_pg['F'][0],
    'ANOVA_p_value': ANOVA_pg['p-unc'][0],
    'kwt_statistic': kwt_pg['H'][0],
    'kwt_p_value': kwt_pg['p-unc'][0],
    'mwut_statistic': mwut_pg['U-val'][0],
    'mwut_p_value': mwut_pg['p-val'][0],
    'swt_statistic': swt_pg['W'][0],
    'swt_p_value': swt_pg['pval'][0],
    'wet_statistic': wet_pg['T'][0],
    'wet_p_value': wet_pg['p-val'][0],
    'z_statistic': zt_pg['T'][0],
    'z_p_value': zt_pg['p-val'][0]
}
```

#**Pandas Library:**

```python
mean_diff = data1.mean() - data2.mean()
std_diff = data1.std() - data2.std()
corr = pd.Series(data1).corr(pd.Series(data2))

result['Pandas'] = {
    'Mean Difference': mean_diff,
    'Standard Deviation Difference': std_diff,
    'Correlation': corr
}
```

#**SymPy Library:**

```python
data1_sympy = sps.density(sps.Normal('X', np.mean(data1), np.std(data1)))(sp.Symbol('x'))
data2_sympy = sps.density(sps.Normal('Y', np.mean(data2), np.std(data2)))(sp.Symbol('y'))
mean1_sympy = sps.E(sps.Normal('X', np.mean(data1), np.std(data1)))
mean2_sympy = sps.E(sps.Normal('Y', np.mean(data2), np.std(data2)))
std1_sympy = sp.sqrt(sps.variance(sps.Normal('X', np.mean(data1), np.std(data1))))
std2_sympy = sp.sqrt(sps.variance(sps.Normal('Y', np.mean(data2), np.std(data2))))

t_sympy = (mean1_sympy - mean2_sympy) / sp.sqrt((std1_sympy**2 / len(data1)) + (std2_sympy**2 / len(data2)))
p_sympy = 2 * (1 - sps.cdf(sps.Normal('Z', 0, 1))(t_sympy))
differences = data1 - data2
n = len(differences)
d = sp.symbols('d:%d' % n)
mean_diff = sum(d) / n
variance_diff = sum([(di - mean_diff)**2 for di in d]) / (n - 1)
std_diff = sp.sqrt(variance_diff)
t_statistic = mean_diff / (std_diff / sp.sqrt(n))
subs = {di: differences[i] for i, di in enumerate(d)}
ptt_t_statistic_sym = t_statistic.evalf(subs=subs)
df = n - 1
ptt_p_value_sym = 2 * (1 - stats.t.cdf(abs(float(ptt_t_statistic_sym)), df))

result['SymPy'] = {
    'Mean Difference': mean1_sympy - mean2_sympy,
    'Standard Deviation Difference': std1_sympy - std2_sympy,
    'itt_statistic': t_sympy,
    'itt_p_value': p_sympy,
    'ptt_statistic': ptt_t_statistic_sym,
    'ptt_p_value': ptt_p_value_sym
}
```

#**Researchpy Library:**

```python
data1_series = pd.Series(data1)
data2_series = pd.Series(data2)

# Welch's t-test
welch_test_rp = rp.ttest(data1_series, data2_series)
itt_rp = rp.ttest(data1_series, data2_series, equal_variances = True, paired = False)
ptt_rp = rp.ttest(data1_series, data2_series, equal_variances = True, paired = True)
wet_rp = rp.ttest(data1_series, data2_series, equal_variances = False, paired = False)
wsrt_rp = rp.ttest(data1_series, data2_series, equal_variances = False, paired = True)

result['Researchpy'] = {
    'wet_statistic': wet_rp[1]['results'][2],
    'wet_p_value': wet_rp[1]['results'][3],
    'itt_statistic': itt_rp[1]['results'][2],
    'itt_p_value': itt_rp[1]['results'][3],
    'ptt_statistic': ptt_rp[1]['results'][2],
    'ptt_p_value': ptt_rp[1]['results'][3],
    'wsrt_statistic': wsrt_rp[1]['results'][0],
    'wsrt_p_value': wsrt_rp[1]['results'][4],
}
```
