Scikit Learn Library:

```python
def Scikit_Learn (X,  clusters):
  return {
      'Silhouette Score': silhouette_score(X, clusters),
      'Davies Boundin Index' :  davies_bouldin_score(X, clusters),
      'Calinski Harabasz Score' : calinski_harabasz_score(X, clusters),
      'WCSS': kmeans.inertia_,
  }
```

Scipy Library:

```python

def silhouette_score_index(X, labels):
    n_clusters = len(np.unique(labels))
    distances = cdist(X, X, 'euclidean')
    silhouette_vals = np.zeros(X.shape[0])

    for i in range(X.shape[0]):
        same_cluster = labels == labels[i]
        a = np.mean(distances[i, same_cluster])
        b = np.min([np.mean(distances[i, labels == other_cluster])
                    for other_cluster in np.unique(labels) if other_cluster != labels[i]])
        silhouette_vals[i] = (b - a) / max(a, b)

    return np.mean(silhouette_vals)



def davies_bouldin_index(X, labels):
    n_clusters = len(np.unique(labels))
    cluster_centroids = np.array([X[labels == i].mean(axis=0) for i in range(n_clusters)])
    S = np.zeros(n_clusters)
    for i in range(n_clusters):
        cluster_points = X[labels == i]
        centroid = cluster_centroids[i]
        S[i] = np.mean(cdist(cluster_points, [centroid]))

    M = cdist(cluster_centroids, cluster_centroids)

    np.fill_diagonal(M, np.inf)

    DB = np.mean([np.max((S[i] + S) / M[i]) for i in range(n_clusters)])

    return DB

import numpy as np
from scipy.spatial.distance import cdist

def calinski_harabasz_index(data, clusters):
    clusters = np.array(clusters)
    n_clusters = len(np.unique(clusters))
    n_samples = len(data)

    if n_clusters == 1 or n_clusters == n_samples:
        return np.nan

    overall_centroid = np.mean(data, axis=0)
    b_dispersion = 0.0
    w_dispersion = 0.0

    for i in np.unique(clusters):

        cluster_data = data[clusters == i]

        cluster_centroid = np.mean(cluster_data, axis=0)

        b_dispersion += len(cluster_data) * np.sum((cluster_centroid - overall_centroid) ** 2)

        w_dispersion += np.sum((cluster_data - cluster_centroid) ** 2)

    ch_index = (b_dispersion / (n_clusters - 1)) / (w_dispersion / (n_samples - n_clusters))

    return ch_index




def Scipy (X, clusters, centers):
  return {
      'Silhouette Score': silhouette_score_index(X, clusters),
      'Davies Boundin Index' :  davies_bouldin_index(X, clusters),
      'Calinski Harabasz Score' : calinski_harabasz_index(X.values, clusters.values),
  }
```

```python
results = {
    'Scikit Learn': Scikit_Learn(X, clusters),
    'Scipy' : Scipy(X, clusters, centers),

}
```
