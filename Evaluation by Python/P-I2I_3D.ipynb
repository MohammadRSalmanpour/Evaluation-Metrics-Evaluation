```python
# 1st Model
import numpy as np
import pandas as pd
from PIL import Image

import os

import nibabel as nib
pred_path = '/content/drive/MyDrive/i2i data for evaluation/pred_SycleGAN3D'
true_path = '/content/drive/MyDrive/i2i data for evaluation/true_MRI'

y_pred = []
y_true = []

for patient in true_path:
    pred = nib.load(os.path.join(true_path, patient[:-4] + '.nii')).get_fdata()
    true = np.load(os.path.join(pred_path, patient[:-4] + '.npz'), allow_pickle=True)['mri']

    y_pred.append(pred.astype(np.float32))
    y_true.append(true.astype(np.float32))

y_true = np.array(y_true)
y_pred = np.array(y_pred)
# Flatten the images for certain metrics
y_true_flat = y_true.flatten()
y_pred_flat = y_pred.flatten()

# Initialize a dictionary to store results
results = {}
from skimage.metrics import mean_squared_error as skimage_mse, peak_signal_noise_ratio as skimage_psnr, structural_similarity as skimage_ssim

# Scikit-Image Metrics
skimage_mse_value = skimage_mse(y_true, y_pred)
skimage_psnr_value = skimage_psnr(y_true, y_pred)

# Sickit Learn
L = float(y_true.max() - y_true.min())
skimage_ssim_value = skimage_ssim(y_true, y_pred, data_range = L)

results['Scikit-Image'] = {
    'MSE': skimage_mse_value,
    'PSNR': skimage_psnr_value,
    'SSIM': skimage_ssim_value
}
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Scikit-Learn Metrics
sklearn_mae = mean_absolute_error(y_true_flat, y_pred_flat)
sklearn_mse = mean_squared_error(y_true_flat, y_pred_flat)
sklearn_rmse = mean_squared_error(y_true_flat, y_pred_flat, squared=False)
sklearn_r2 = r2_score(y_true_flat, y_pred_flat)

results['Scikit-Learn'] = {
    'MAE': sklearn_mae,
    'MSE': sklearn_mse,
    'RMSE': sklearn_rmse,
    'R²': sklearn_r2
}


from statsmodels.tools.eval_measures import mse, rmse
import statsmodels.api as sm

# Statsmodels Metrics
model = sm.OLS(y_true_flat, sm.add_constant(y_pred_flat)).fit()
sm_mae = mean_absolute_error(y_true_flat, y_pred_flat)
sm_mse = mse(y_true_flat, y_pred_flat)
sm_rmse = rmse(y_true_flat, y_pred_flat)
sm_r2 = model.rsquared

results['Statsmodels'] = {
    'R²': sm_r2,
    'MAE': sm_mae,
    'MSE': sm_mse,
    'RMSE': sm_rmse
}

# Pandas setting to show more columns
pd.set_option('display.max_columns', None)  # Showing all columns
pd.set_option('display.max_colwidth', None)  # Setting all columns wide

# Convert results to a DataFrame
results_df = pd.DataFrame(results).T
results_df.index.name = 'Library'  # Set index name for better readability
results_df.reset_index(inplace=True)  # Reset index to make 'Library' a column
results_df = results_df.rename_axis(None, axis=1)  # Remove axis name for columns
print(results_df)
results_df.to_excel('/content/drive/MyDrive/i2i data for evaluation/metrics_i2i.xlsx', index = False)
```

```python
# i2i with batch 10 - 120 sample 3D

import os
import numpy as np
import pandas as pd
import nibabel as nib
from skimage.metrics import mean_squared_error as skimage_mse, peak_signal_noise_ratio as skimage_psnr, structural_similarity as skimage_ssim
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from statsmodels.tools.eval_measures import mse, rmse
import statsmodels.api as sm

# مسیرهای دو پوشه
pred_path = '/content/drive/MyDrive/Mr-Alizadeh/i2i data for evaluation/pred_SycleGAN3D'
true_path = '/content/drive/MyDrive/Mr-Alizadeh/i2i data for evaluation/true_MRI'

# پارامتر batch size
BATCH_SIZE = 10

# دریافت لیست فایل‌های true با پسوند .nii
true_files = [f for f in os.listdir(true_path) if f.endswith('.nii')]

# تابعی برای پردازش هر batch از فایل‌ها
def process_batch(batch_files):
    y_pred = []
    y_true = []

    for patient in batch_files:
        pred_file = os.path.join(pred_path, patient[:-4] + '.npz')
        true_file = os.path.join(true_path, patient)

        pred = np.load(pred_file, allow_pickle=True)['mri']
        true = nib.load(true_file).get_fdata()

        y_pred.append(pred.astype(np.float32))
        y_true.append(true.astype(np.float32))

    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    return y_true, y_pred

# متغیر نتایج نهایی
results = {
    'Scikit-Image': {'MSE': 0, 'PSNR': 0, 'SSIM': 0},
    'Scikit-Learn': {'MAE': 0, 'MSE': 0, 'RMSE': 0, 'R²': 0},
    'Statsmodels': {'R²': 0, 'MAE': 0, 'MSE': 0, 'RMSE': 0}
}

# پردازش داده‌ها به صورت batch
for i in range(0, len(true_files), BATCH_SIZE):
    batch_files = true_files[i:i + BATCH_SIZE]
    y_true, y_pred = process_batch(batch_files)

    # محاسبه معیارها
    y_true_flat = y_true.flatten()
    y_pred_flat = y_pred.flatten()

    # Scikit-Image Metrics
    results['Scikit-Image']['MSE'] += skimage_mse(y_true, y_pred)
    results['Scikit-Image']['PSNR'] += skimage_psnr(y_true, y_pred)
    L = float(y_true.max() - y_true.min())
    results['Scikit-Image']['SSIM'] += skimage_ssim(y_true, y_pred, data_range=L)

    # Scikit-Learn Metrics
    results['Scikit-Learn']['MAE'] += mean_absolute_error(y_true_flat, y_pred_flat)
    results['Scikit-Learn']['MSE'] += mean_squared_error(y_true_flat, y_pred_flat)
    results['Scikit-Learn']['RMSE'] += mean_squared_error(y_true_flat, y_pred_flat, squared=False)
    results['Scikit-Learn']['R²'] += r2_score(y_true_flat, y_pred_flat)

    # Statsmodels Metrics
    model = sm.OLS(y_true_flat, sm.add_constant(y_pred_flat)).fit()
    results['Statsmodels']['R²'] += model.rsquared
    results['Statsmodels']['MAE'] += mean_absolute_error(y_true_flat, y_pred_flat)
    results['Statsmodels']['MSE'] += mse(y_true_flat, y_pred_flat)
    results['Statsmodels']['RMSE'] += rmse(y_true_flat, y_pred_flat)

# محاسبه میانگین معیارها
num_batches = len(true_files) / BATCH_SIZE
for key in results:
    for metric in results[key]:
        results[key][metric] /= num_batches

# تنظیمات pandas
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)

# تبدیل نتایج به DataFrame
results_df = pd.DataFrame(results).T
results_df.index.name = 'Library'
results_df.reset_index(inplace=True)
results_df = results_df.rename_axis(None, axis=1)

# نمایش و ذخیره نتایج
print(results_df)
results_df.to_excel('/content/drive/MyDrive/Loss Function Evaluation/Code and Results/Segmentation/Binary_Label Segmentation/Results/metrics_i2i2.xlsx', index=False)
```

```python
# The best code

import os
import numpy as np
import pandas as pd
import nibabel as nib
from skimage.metrics import mean_squared_error as skimage_mse, peak_signal_noise_ratio as skimage_psnr, structural_similarity as skimage_ssim
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from statsmodels.tools.eval_measures import mse, rmse
import statsmodels.api as sm
import tensorflow as tf
import torch
import torchmetrics
import piq

# مسیرهای دو پوشه
pred_path = '/content/drive/MyDrive/Mr-Alizadeh/i2i data for evaluation/pred_SycleGAN3D'
true_path = '/content/drive/MyDrive/Mr-Alizadeh/i2i data for evaluation/true_MRI'

# پارامتر batch size
BATCH_SIZE = 10

true_files = [f for f in os.listdir(true_path) if f.endswith('.nii')]

def process_batch(batch_files):
    y_pred, y_true = [], []
    for patient in batch_files:
        pred_file = os.path.join(pred_path, patient[:-4] + '.npz')
        true_file = os.path.join(true_path, patient)

        pred = np.load(pred_file, allow_pickle=True)['mri']
        true = nib.load(true_file).get_fdata()

        y_pred.append(pred.astype(np.float32))
        y_true.append(true.astype(np.float32))

    return np.array(y_true), np.array(y_pred)

batch_results = {
    'Scikit-Image': {'MSE': [], 'PSNR': [], 'SSIM': []},
    'Scikit-Learn': {'MAE': [], 'MSE': [], 'RMSE': [], 'R²': []},
    'Statsmodels': {'R²': [], 'MAE': [], 'MSE': [], 'RMSE': []},
    'TensorFlow': {'PSNR': [], 'SSIM': []},
    'PyTorch': {'MAE': [], 'MSE': [], 'RMSE': [], 'PSNR': [], 'SSIM': []}
}

for i in range(0, len(true_files), BATCH_SIZE):
    batch_files = true_files[i:i + BATCH_SIZE]
    y_true, y_pred = process_batch(batch_files)

    y_true_flat = y_true.flatten()
    y_pred_flat = y_pred.flatten()

    # Scikit-Image Metrics
    batch_results['Scikit-Image']['MSE'].append(skimage_mse(y_true, y_pred))
    batch_results['Scikit-Image']['PSNR'].append(skimage_psnr(y_true, y_pred))
    L = float(y_true.max() - y_true.min())
    batch_results['Scikit-Image']['SSIM'].append(skimage_ssim(y_true, y_pred, data_range=L))

    # Scikit-Learn Metrics
    batch_results['Scikit-Learn']['MAE'].append(mean_absolute_error(y_true_flat, y_pred_flat))
    batch_results['Scikit-Learn']['MSE'].append(mean_squared_error(y_true_flat, y_pred_flat))
    batch_results['Scikit-Learn']['RMSE'].append(mean_squared_error(y_true_flat, y_pred_flat, squared=False))
    batch_results['Scikit-Learn']['R²'].append(r2_score(y_true_flat, y_pred_flat))

    # Statsmodels Metrics
    model = sm.OLS(y_true_flat, sm.add_constant(y_pred_flat)).fit()
    batch_results['Statsmodels']['R²'].append(model.rsquared)
    batch_results['Statsmodels']['MAE'].append(mean_absolute_error(y_true_flat, y_pred_flat))
    batch_results['Statsmodels']['MSE'].append(mse(y_true_flat, y_pred_flat))
    batch_results['Statsmodels']['RMSE'].append(rmse(y_true_flat, y_pred_flat))

    # TensorFlow Metrics
    y_true_tf = tf.convert_to_tensor(y_true)
    y_pred_tf = tf.convert_to_tensor(y_pred)
    batch_results['TensorFlow']['PSNR'].append(tf.image.psnr(y_true_tf, y_pred_tf, max_val=1.0).numpy())
    batch_results['TensorFlow']['SSIM'].append(tf.image.ssim(y_true_tf, y_pred_tf, max_val=1.0).numpy())

    # PyTorch Metrics
    y_true_torch = torch.tensor(y_true)
    y_pred_torch = torch.tensor(y_pred)
    mae_metric = torchmetrics.MeanAbsoluteError()
    mse_metric = torchmetrics.MeanSquaredError()
    rmse_metric = torchmetrics.MeanSquaredError(squared=False)
    mae_pt = mae_metric(y_pred_torch, y_true_torch).item()
    mse_pt = mse_metric(y_pred_torch, y_true_torch).item()
    rmse_pt = rmse_metric(y_pred_torch, y_true_torch).item()
    y_true_torch = (y_true_torch - y_true_torch.min()) / (y_true_torch.max() - y_true_torch.min())
    y_pred_torch = (y_pred_torch - y_pred_torch.min()) / (y_pred_torch.max() - y_pred_torch.min())

    batch_results['PyTorch']['MAE'].append(mae_pt)
    batch_results['PyTorch']['MSE'].append(mse_pt)
    batch_results['PyTorch']['RMSE'].append(rmse_pt)
    batch_results['PyTorch']['PSNR'].append(piq.psnr(y_pred_torch, y_true_torch, data_range=1.0).item())
    batch_results['PyTorch']['SSIM'].append(piq.ssim(y_pred_torch, y_true_torch, data_range=1.0).item())

# Compute the final average
results = {library: {metric: np.mean(values) for metric, values in metrics.items()}
           for library, metrics in batch_results.items()}

results_df = pd.DataFrame(results).T.reset_index().rename_axis(None, axis=1)
print(results_df)

results_df.to_excel('/content/drive/MyDrive/Loss Function Evaluation/Code and Results/I2I Translation/Results/metrics_i2i5.xlsx', index=False)
```
